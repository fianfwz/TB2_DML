```python
# IMPORT LIBRARY
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from wordcloud import WordCloud
import re
import nltk
from nltk.tokenize import word_tokenize
import warnings
from datetime import datetime

# IGNORE WARNING
warnings.filterwarnings('ignore')

# DOWNLOAD NLTK RESOURCES YANG DIBUTUHKAN
nltk.download('punkt', quiet=True)

# ==================================================
# A. ANALISIS DATA KOSONG (MISSING VALUE)
# ==================================================

# Upload File
from google.colab import files
uploaded = files.upload()
```

```python
# 1. Buka dataset dan tampilkan 5 baris pertama
df = pd.read_csv('dataset_dummy_komentar_korupsi_timah.csv')
print("="*80)
print("1. 5 BARIS PERTAMA DATASET:")
print("="*80)
print(df.head())
print("\n")
```

```python
# 2. Jumlah data null per kolom
print("="*80)
print("2. JUMLAH DATA NULL PER KOLOM:")
print("="*80)
null_counts = df.isnull().sum()
print(null_counts)
print("\n")

# 3. Visualisasi missing value dalam bentuk bar chart
print("="*80)
print("3. VISUALISASI MISSING VALUE (BAR CHART):")
print("="*80)

# Hitung jumlah missing value per kolom
missing_values = df.isnull().sum()

# Filter hanya kolom yang memiliki missing value
missing_values = missing_values[missing_values > 0]

# Buat visualisasi
plt.figure(figsize=(10, 6))
missing_values.plot(kind='bar', color='#87cefa')
plt.title('Jumlah Missing Value per Kolom')
plt.ylabel('Jumlah Missing Value')
plt.xlabel('Kolom')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
# 4. Membuat dua versi dataset dengan penanganan missing value yang lebih baik
print("="*80)
print("4. VERSI DATASET:")
print("="*80)

# Versi original
df_original = df.copy()

# Tampilkan 25 data sebelum penanganan missing value
print("\n25 DATA SEBELUM PENANGANAN MISSING VALUE:")
print(df_original[['email', 'akun', 'tanggal', 'komentar']].head(25).to_string())

# Versi cleaned - handle missing values dengan logika yang lebih baik
df_cleaned = df.copy()

# Fungsi untuk mengisi missing value dengan lebih baik
def fill_missing_values(row):
    # Jika akun kosong tapi email ada
    if pd.isna(row['akun']) and pd.notna(row['email']):
        # Ambil bagian sebelum @ dari email sebagai akun
        row['akun'] = row['email'].split('@')[0]

    # Jika email kosong tapi akun ada
    elif pd.isna(row['email']) and pd.notna(row['akun']):
        # Gunakan akun + @gmail.com untuk email
        row['email'] = f"{row['akun']}@gmail.com"

    return row

# Terapkan fungsi ke setiap baris
df_cleaned = df_cleaned.apply(fill_missing_values, axis=1)

# Hapus baris tanpa komentar
df_cleaned = df_cleaned.dropna(subset=['komentar']).reset_index(drop=True)

# TAMBAHAN REVISI: Hapus baris dimana email dan akun masih kosong setelah diisi
rows_before = len(df_cleaned)
df_cleaned = df_cleaned[~((df_cleaned['email'].isna()) & (df_cleaned['akun'].isna()))]
rows_removed = rows_before - len(df_cleaned)

# Untuk baris yang masih memiliki akun kosong (karena email juga kosong), isi dengan 'unknown'
df_cleaned['akun'] = df_cleaned['akun'].fillna('unknown')
df_cleaned['email'] = df_cleaned['email'].fillna('unknown@example.com')

# PENANGANAN TANGGAL YANG KOSONG
# Konversi kolom tanggal ke datetime
df_cleaned['tanggal'] = pd.to_datetime(df_cleaned['tanggal'], errors='coerce')

# Hitung rata-rata tanggal (dari nilai yang tersedia)
mean_date = df_cleaned['tanggal'].mean()

# Isi missing value dengan rata-rata tanggal
df_cleaned['tanggal'] = df_cleaned['tanggal'].fillna(mean_date)

print(f"\nJumlah data original: {len(df_original)}")
print(f"Jumlah data cleaned: {len(df_cleaned)}")
print(f"Baris dihapus (email dan akun keduanya kosong): {rows_removed}")
print(f"Tanggal rata-rata yang digunakan untuk mengisi missing value: {mean_date.date()}")

# Tampilkan ringkasan missing value
print("\nSTATUS MISSING VALUE SETELAH PENANGANAN:")
print(df_cleaned.isnull().sum())

# Tampilkan 25 data setelah penanganan missing value
print("\n25 DATA SETELAH PENANGANAN MISSING VALUE:")
print(df_cleaned[['email', 'akun', 'tanggal', 'komentar']].head(25).to_string())

print("\nPERBANDINGAN DATA SEBELUM DAN SESUDAH:")
# Ambil index acak untuk perbandingan
sample_indices = df_cleaned.index[:5].tolist()  # Ambil 5 index pertama

for idx in sample_indices:
    original_row = df_original.loc[idx] if idx in df_original.index else None
    cleaned_row = df_cleaned.loc[idx]

    print(f"\nData index {idx}:")
    print("SEBELUM:")
    print(f"Email: {original_row['email'] if original_row is not None else 'Tidak ada di data original'}")
    print(f"Akun: {original_row['akun'] if original_row is not None else 'Tidak ada di data original'}")
    print(f"Tanggal: {original_row['tanggal'] if original_row is not None else 'Tidak ada di data original'}")
    print("\nSESUDAH:")
    print(f"Email: {cleaned_row['email']}")
    print(f"Akun: {cleaned_row['akun']}")
    print(f"Tanggal: {cleaned_row['tanggal'].date()}")
    print("-"*50)
```

```python
# ==================================================
# B. PEMBENTUKAN CLUSTER KOMENTAR (TEXT CLUSTERING)
# ==================================================

# 1. Preprocessing teks
print("="*80)
print("B.1 PREPROCESSING TEKS:")
print("="*80)
factory = StemmerFactory()
stemmer = factory.create_stemmer()
stop_factory = StopWordRemoverFactory()
stopwords = stop_factory.get_stop_words()

# Fungsi tokenizer aman
def safe_tokenize(text):
    try:
        return word_tokenize(text)
    except LookupError:
        # Fallback: tokenizer berbasis regex
        return re.findall(r'\b\w+\b', text)

def preprocess_text(text):
    if pd.isna(text) or not isinstance(text, str):
        return ''

    # Lowercase
    text = text.lower()

    # Hapus tanda baca dan karakter khusus
    text = re.sub(r'[^\w\s]', ' ', text)

    # Hapus angka
    text = re.sub(r'\d+', '', text)

    # Hapus spasi berlebih
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenisasi dengan fallback
    tokens = safe_tokenize(text)

    # Stopword removal dan stemming
    tokens = [
        stemmer.stem(word)
        for word in tokens
        if word not in stopwords and len(word) > 2
    ]

    return ' '.join(tokens) if tokens else ''

# Terapkan preprocessing
df_cleaned['cleaned_komentar'] = df_cleaned['komentar'].apply(preprocess_text)

# Hapus komentar yang kosong setelah preprocessing
df_cleaned = df_cleaned[df_cleaned['cleaned_komentar'] != ''].reset_index(drop=True)

# Tampilkan 5 contoh hasil preprocessing
print("\nContoh Hasil Preprocessing (5 Sample):")
for i in range(5):
    print(f"\nOriginal ({i}):", df_cleaned['komentar'].iloc[i])
    print(f"Cleaned  ({i}):", df_cleaned['cleaned_komentar'].iloc[i])
```

```python
# 2. TF-IDF Vectorizer dengan Visualisasi
print("="*80)
print("B.2 TF-IDF VECTORIZER DENGAN VISUALISASI:")
print("="*80)

# Inisialisasi TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(
    max_features=1000,
    ngram_range=(1, 2),
    min_df=5,
    max_df=0.8
)

# Fit dan transform data
tfidf_matrix = tfidf_vectorizer.fit_transform(df_cleaned['cleaned_komentar'])
print(f"Dimensi TF-IDF Matrix: {tfidf_matrix.shape}")
print(f"Jumlah fitur (terms): {len(tfidf_vectorizer.get_feature_names_out())}")

# 1. Visualisasi Top Terms berdasarkan TF-IDF Score
print("\n1. TOP 20 TERMS BERDASARKAN TF-IDF SCORE:")

# Hitung rata-rata TF-IDF score per term
mean_tfidf = np.asarray(tfidf_matrix.mean(axis=0)).ravel()
top_indices = mean_tfidf.argsort()[::-1][:20]  # Indeks 20 terms teratas
top_terms = tfidf_vectorizer.get_feature_names_out()[top_indices]
top_scores = mean_tfidf[top_indices]

plt.figure(figsize=(12, 8))
sns.barplot(x=top_scores, y=top_terms, palette="viridis")
plt.title('Top 20 Terms Berdasarkan Rata-Rata TF-IDF Score')
plt.xlabel('Rata-Rata TF-IDF Score')
plt.ylabel('Term')
plt.tight_layout()
plt.show()
```

```python
# Pastikan optimal_clusters = 3
optimal_clusters = 3  # Tambahkan ini jika belum ada

# 4. Melatih model K-Means dengan 3 cluster
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=optimal_clusters, random_state=42, n_init=10)
kmeans.fit(tfidf_matrix)
df_cleaned['cluster'] = kmeans.labels_

# 5. Analisis hasil clustering
print("="*80)
print(f"HASIL CLUSTERING ({optimal_clusters} CLUSTER):")  # Gunakan variabel optimal_clusters
print("="*80)

# Distribusi cluster
cluster_distribution = df_cleaned['cluster'].value_counts().sort_index()
print("\nJUMLAH KOMENTAR PER CLUSTER:")
print(cluster_distribution)

# Fungsi untuk mendapatkan karakteristik cluster
def get_cluster_characteristics(cluster_id, top_n=10):
    cluster_mask = (df_cleaned['cluster'] == cluster_id).values
    cluster_features = tfidf_matrix[cluster_mask]

    avg_tfidf = np.asarray(cluster_features.mean(axis=0)).ravel()
    top_indices = avg_tfidf.argsort()[-top_n:][::-1]
    features = tfidf_vectorizer.get_feature_names_out()

    return [features[i] for i in top_indices]

# Analisis tiap cluster
print("\nKARAKTERISTIK TIAP CLUSTER:")
for cluster_id in range(optimal_clusters):  # Gunakan optimal_clusters di loop
    characteristics = get_cluster_characteristics(cluster_id)
    print(f"\nCLUSTER {cluster_id} (Jumlah: {cluster_distribution[cluster_id]}):")
    print("  - Kata Kunci Utama:", ', '.join(characteristics))

    # Contoh komentar
    print("  - Contoh Komentar:")
    cluster_samples = df_cleaned[df_cleaned['cluster'] == cluster_id]['komentar'].sample(3, random_state=42)
    for i, comment in enumerate(cluster_samples, 1):
        truncated = (comment[:100] + '...') if len(comment) > 100 else comment
        print(f"    {i}. {truncated}")

# 6. Visualisasi dengan PCA
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

print("\n" + "="*80)
print(f"VISUALISASI {optimal_clusters} CLUSTER (PCA):")  # Gunakan optimal_clusters
print("="*80)

# Reduksi dimensi
pca = PCA(n_components=2, random_state=42)
reduced_features = pca.fit_transform(tfidf_matrix.toarray())

# Plot
plt.figure(figsize=(12, 8))
scatter = plt.scatter(
    reduced_features[:, 0],
    reduced_features[:, 1],
    c=df_cleaned['cluster'],
    cmap='viridis',
    alpha=0.7,
    s=50
)

plt.colorbar(scatter, ticks=range(optimal_clusters), label='Cluster')  # Gunakan optimal_clusters
plt.title(f'Visualisasi {optimal_clusters} Cluster Komentar (Reduksi PCA)')  # Gunakan optimal_clusters
plt.xlabel('PCA Komponen 1')
plt.ylabel('PCA Komponen 2')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# 7. Visualisasi WordCloud untuk Komentar Negatif
print("="*80)
print("B.7 WORDCLOUD KOMENTAR NEGATIF:")
print("="*80)

# Identifikasi komentar negatif berdasarkan kata kunci negatif
negative_keywords = [
    'korupsi', 'pencuri', 'hukum', 'pidana', 'kejahatan',
    'merugikan', 'tindak', 'kriminal', 'skandal', 'penjara'
]

# Filter komentar yang mengandung kata kunci negatif
def is_negative(text):
    return any(keyword in text for keyword in negative_keywords)

df_cleaned['is_negative'] = df_cleaned['cleaned_komentar'].apply(is_negative)
negative_comments = df_cleaned[df_cleaned['is_negative']]

# Gabungkan semua teks komentar negatif
negative_text = ' '.join(negative_comments['cleaned_komentar'])

# Buat WordCloud
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    colormap='Reds',  # Warna merah untuk sentimen negatif
    max_words=100,
    stopwords=None  # Tambahkan stopwords khusus jika perlu
).generate(negative_text)

# Tampilkan WordCloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('WordCloud - Komentar Negatif', fontsize=16)
plt.tight_layout(pad=0)
plt.show()

# Tampilkan statistik
print(f"Jumlah komentar negatif: {len(negative_comments)}")
print("\nContoh komentar negatif:")
for i, comment in enumerate(negative_comments['komentar'].head(3), 1):
    print(f"{i}. {comment[:150]}...")
```

```python
# 3. Clustering dengan K-Means
print("="*80)
print("B.3 CLUSTERING DENGAN K-MEANS:")
print("="*80)

# 4. Menentukan jumlah cluster optimal
print("Menentukan Jumlah Cluster Optimal:")
wcss = []  # Untuk Elbow Method
silhouette_scores = []  # Untuk Silhouette Score
cluster_range = range(2, 8)

for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(tfidf_matrix)
    wcss.append(kmeans.inertia_)

    # Silhouette Score hanya untuk k > 1
    if k > 1:
        score = silhouette_score(tfidf_matrix, kmeans.labels_)
        silhouette_scores.append(score)
        print(f"  - k={k}: Silhouette Score = {score:.4f}")

# Plot Elbow Method
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(cluster_range, wcss, 'bo-')
plt.xlabel('Jumlah Cluster')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.grid(True)

# Plot Silhouette Score
plt.subplot(1, 2, 2)
plt.plot(range(2, 8), silhouette_scores, 'go-')
plt.xlabel('Jumlah Cluster')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score')
plt.grid(True)

plt.tight_layout()
plt.show()
```
